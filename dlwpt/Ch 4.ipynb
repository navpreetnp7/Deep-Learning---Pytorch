```python
path = 'data/ch4/pablo.txt'
with open(path,encoding='utf8') as f:
    text = f.read()
lines = text.split('\n')    
```

```python
def clean_words(ipstr):
    punctuations = '.,;:"!?"_-'
    word_list = ipstr.lower().replace('\n',' ').split(' ')
    word_list = [word.strip(punctuations) for word in word_list]
    return word_list
```

```python
word_list = sorted(set(clean_words(text)))
word2idx_dict = {word:i for (i,word) in enumerate(word_list)}
```

```python
len(word2idx_dict), word2idx_dict['pablo']
```

```python
words = clean_words(text)
# one hot encoding
words_t = torch.zeros(len(words),len(word2idx_dict))
for i,word in enumerate(words):
    word_idx = word2idx_dict[word]
    words_t[word_idx] = 1
    print (i,word_idx,word)
```
